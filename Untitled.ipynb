````python
#complete answer three as a function
def answer_three():
    GDP = pd.read_csv('C:/Users/limpertl/Documents/Jupyter Notebooks/DataFrame_Assignment_Week_3/world_bank.csv',
                      header=4)
    GDP = GDP [['Country Name', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
    GDP.set_index(['Country Name'], inplace=True)
    avgGDP = pd.Series(GDP.mean(axis=1))
    avgGDP.sort_values(axis=0, ascending=False,inplace=True,)
    avgGDP = avgGDP[:15]
    return avgGDP
````
